```python
import re
```

```python
def remove_tags(text):
    seperated_head = re.sub('<h3>', '\n', text)
    no_tags = re.sub(r'<[^>]+>', '', seperated_head)
#     space_after_punc = re.sub('[.:?!-",]\w' ,'[.:?!-",]\s\w', no_tags)
    no_spaces = re.sub('  +', ' ', no_tags)
    return no_spaces
```

```python
print(remove_tags(text))
```

## 4. Pymorphy vs Mystem (2 балла)

Лемматизируйте очищенный текст с помощью mystem и pymorphy. Сравните результаты: есть ли разница, что лучше. 

Для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат.

```python
from pymystem3 import Mystem
from pymorphy2 import MorphAnalyzer
```

### Pymorhy

```python
mystem = Mystem()
tokens = mystem.analyze(text)

morph = MorphAnalyzer(lang='ru') 
prepared_tokens = [tokens[i]['text'] for i in range(len(tokens)) if tokens[i].get('analysis')] 
next_words_analized = [morph.parse(token) for token in prepared_tokens]
text_pymorphy = [next_words_analized[i][0].normal_form for i in range(len(next_words_analized))]

print(len(text_pymorphy))
```

### Mystem

```python
mystem = Mystem()
tokens = mystem.analyze(text)
text_mystem = [parse['analysis'][0]['lex'] for parse in tokens if parse.get('analysis')]
print(len(text_mystem))
```

### Сравнение

```python
unoverlapped = 0
for i in range(len(text_mystem)):
    if text_mystem[i] != text_pymorphy[i]:
        unoverlapped += 1
        print(f'{i}. Исходный: {prepared_tokens[i]}, PyMorphy: {text_pymorphy[i]}, MyStem: {text_mystem[i]}')

print(f'- - - - - - - - - - - -')
print(f'Всего несовпадений: {unoverlapped}')
print(f'В процентах: {unoverlapped/len(text_mystem)*100:.2f}%')
```

Справляются они почти что одинаково. Но есть и различия (в основном в том, какую форму считать нулевой):
1. PyMorhy чувствует разницу между совершенным и несовершенным видами, MyStem лемматизирует всякую форму глагола в несовершенный вид (примеры № 671, 607, 553, 547, 472 и тд)
2. PyMorhy лемматизирует сравнительную форму прилагательных и наречий как прилагательное, MyStem - часто приводит к наречию, а не к прилагательному (примеры №73,76, 69, 508 и тд)
3. Указательные местоимения PyMorphy приводит к считающейся, но не грамматической нулевой форме (этом -> это, при нулевой форме этот) (примеры 539, 433 и тд)
4. У PyMorhy есть ё, а это чего-то стоит (примеры №542, 526, 496 и тд)

Явные ошибки, а не разночтения:
1. PyMorphy 
    1. Лемматизирует есть как есть, а не как быть (есть - форма тертьего лица настоящего времени глагола быть), вероятно, так как эта форма более частотная (примеры №635, 488 и др). Занятно, но е в т.е. PyMorphy лемматизирует как быть, что скорее верно, а вот т в т.е. - неверно (пример 210-211)
    2. Придумывает формы незнакомых (несловарный) слов в зависимости от их "внешнего вида" (примеры №644, 549,80)
2. MyStem
    1. Выбирает более частотные варианты от чего иногда не попадает в контекст (примеры 450, )
    2. Не дружит с сокращениями (пример 210-211, 545, 533 и др)
    3. Иногда неверно указывает нулевую форму указательных местоимений (пример 215)
